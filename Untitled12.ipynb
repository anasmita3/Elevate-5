```python
# -----------------------------------------------------------
# Exploratory Data Analysis ‚Äì Mobile Price Dataset
# -----------------------------------------------------------
# Dependencies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid", context="notebook")
pd.set_option("display.max_columns", None)

# 1. Load the data
df = pd.read_csv("/content/dataset1.csv")
print("Shape :", df.shape)
display(df.head())

# 2. Quick structural checks
print("\n--- .info() ---")
df.info()

print("\n--- .describe() ---")
display(df.describe().T)

print("\n--- Missing‚Äêvalue count ---")
display(df.isnull().sum())

# 3. Univariate distributions
num_cols = ["battery_power", "ram", "px_height", "px_width"]
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, col in zip(axes.flatten(), num_cols):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f"Distribution of {col}")
plt.tight_layout()
plt.show()

# 4. Box‚Äëand‚Äëwhisker: RAM vs. price tier
plt.figure(figsize=(6, 4))
sns.boxplot(x="price_range", y="ram", data=df)
plt.title("RAM by Price Range")
plt.show()

# 5. Correlation heatmap (numeric features)
plt.figure(figsize=(10, 8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# 6. Pairplot of the most informative features
subset = df[["price_range", "ram", "battery_power", "px_height", "px_width"]]
sns.pairplot(subset, hue="price_range", diag_kind="hist")
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```
